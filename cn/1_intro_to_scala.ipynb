# 理解 Scala
Scala 是一种支持常见编程范式的编程语言。我们选择使用它，原因如下：

- 它是一种很好的宿主语言，适合嵌入 DSL（领域特定语言）。
- 它拥有一个强大而优雅的库，用于操作各种数据集合。
- 它具有严格的类型系统，有助于在开发周期的早期（即编译时）捕获大量错误。
- 它具有强大的函数表达和传递能力。
- Chi*p*el、Chi*j*el 和 Chi*c*el 的发音远不如 Chisel 好听。

所有这些要点将在我们稍后讨论 Chisel 时变得明显，但现在，我们将专注于阅读和编写 Scala 代码的基础知识。

# 条件语句

Scala 实现条件语句的方式与其他编程语言类似。

````scala
// A simple conditional; by the way, this is a comment
if (numberOfKittens > kittensPerHouse) { 
    println("Too many kittens!!!") 
}
// The braces are not required when all branches are one liners. However, the 
// Scala Style Guide prefers brace omission only if an "else" clause is included.
// (Preferably not this, even though it compiles...)
if (numberOfKittens > kittensPerHouse) 
    println("Too many kittens!!!")

// ifs have else clauses, of course
// This is where you can omit braces!
if (done) 
    println("we are done")
else 
    numberOfKittens += 1

// And else ifs
// For style, keep braces because not all branches are one liners. 
if (done) {
    println("we are done")
}
else if (numberOfKittens < kittensPerHouse) {
    println("more kittens!")
    numberOfKittens += 1
}
else {
    done = true
}
````

但在 Scala 中，`if` 条件语句会返回一个值。这个值是什么？它由所选分支的最后一行给出。这个特性非常强大，特别是在初始化函数和类中的值时。它看起来像这样：

````scala
val likelyCharactersSet = if (alphabet.length == 26)
    "english"
else 
    "not english"

println(likelyCharactersSet)
````

我们创建了一个常量 *likelyCharactersSet*，其值在运行时根据条件确定。

# 方法（函数）

方法使用关键字 `def` 定义。在这里，我们也会不拘泥于术语地称它们为函数。函数参数（或形参）在以逗号分隔的列表中指定，列表中包含参数名、类型以及可选的默认值。为清晰起见，应该指定返回类型。

没有任何参数的 Scala 函数不需要空括号。当一个类的成员因为与引用它相关联的某些计算而变成函数时，这对开发者来说通常很方便。按照惯例，没有副作用的无参函数（即调用它们不会改变任何东西，只是返回一个值）不使用括号，而有副作用的函数（可能会改变类变量或打印东西）应该要求括号。

## 简单声明

````scala
// Simple scaling function with an input argument, e.g., times2(3) returns 6
// Curly braces can be omitted for short one-line functions.
def times2(x: Int): Int = 2 * x

// More complicated function
def distance(x: Int, y: Int, returnPositive: Boolean): Int = {
    val xy = x * y
    if (returnPositive) xy.abs else -xy.abs
}
````

## 函数重载
相同的函数名可以以多种方式使用。参数及其类型确定了一个签名，使编译器能够确定应该调用哪个版本的函数。不过，应该避免使用重载函数。

````scala
// Overloaded function
def times2(x: Int): Int = 2 * x
def times2(x: String): Int = 2 * x.toInt

times2(5)
times2("7")
````

## 递归和嵌套函数
花括号定义代码作用域。在函数的作用域内可以存在更多函数或递归函数调用。在某个作用域中定义的函数只能在该作用域内访问。

````scala
/** Prints a triangle made of "X"s
  * This is another style of comment
  */
def asciiTriangle(rows: Int) {
    
    // This is cute: multiplying "X" makes a string with many copies of "X"
    def printRow(columns: Int): Unit = println("X" * columns)
    
    if(rows > 0) {
        printRow(rows)
        asciiTriangle(rows - 1) // Here is the recursive call
    }
}

// printRow(1) // This would not work, since we're calling printRow outside its scope
asciiTriangle(6)
````

# 列表（Lists）
Scala 实现了各种聚合或序列对象。列表很像数组，但支持追加和提取等额外操作。

````scala
val x = 7
val y = 14
val list1 = List(1, 2, 3)
val list2 = x :: y :: y :: Nil       // An alternate notation for assembling a list

val list3 = list1 ++ list2           // Appends the second list to the first list
val m = list2.length
val s = list2.size

val headOfList = list1.head          // Gets the first element of the list
val restOfList = list1.tail          // Get a new list with first element removed

val third = list1(2)                 // Gets the third element of a list (0-indexed)
````

# `for` 语句

Scala 有一个 `for` 语句，其工作方式类似于传统的 for 语句。你可以遍历一个值范围。

````scala
for (i <- 0 to 7) { print(i + " ") }
println()
````

使用 `until` 代替 `to` 可以从 0 遍历到 6（不包含 7）。

````scala
for (i <- 0 until 7) { print(i + " ") }
println()
````

添加 `by` 可以按固定步长递增。以下代码打印出 0 到 10 之间的偶数。

````scala
for(i <- 0 to 10 by 2) { print(i + " ") }
println()
````

如果你有某种集合，想要访问所有元素，可以使用 `for` 作为*迭代器*，就像 Java 和 Python 中那样。这里我们创建一个包含 4 个随机整数的列表，然后对它们求和。

````scala
val randomList = List(scala.util.Random.nextInt(), scala.util.Random.nextInt(), scala.util.Random.nextInt(), scala.util.Random.nextInt())
var listSum = 0
for (value <- randomList) {
  listSum += value
}
println("sum is " + listSum)
````

Scala 的 `for` 还有更多技巧。对于各种传统迭代需求，它都能直观地工作，但它不一定总是最方便的选择。像对数组元素求和这样的操作，通常使用一组丰富的函数（称为*推导式*）更容易完成，这些函数在许多不同类型的集合上都可用。后续模块将详细介绍 `for` 及其相关工具。

# Scala 是一种面向对象语言
Scala 是面向对象的，理解这一点对于充分利用 Scala 和 Chisel 非常重要。请注意，毫无疑问，描述这一切的方式不止一种。
1. 变量是对象。
1. Scala 中 `val` 声明的常量也是对象。
1. 甚至字面量值也是对象。
1. 甚至函数本身也是对象。稍后会详细讨论。
1. 对象是类的实例。
  1. 事实上，在 Scala 中几乎在所有重要的方面，*面向对象*中的*对象*都被称为实例。
1. 在定义类时，程序员指定：
  1. 与类关联的数据（`val`、`var`）。
  1. 类的实例可以执行的操作，称为方法或函数。
1. 类可以扩展其他类。
  1. 被扩展的类是超类；扩展者是子类。
  1. 在这种情况下，子类继承超类的数据和方法。
  1. 有许多有用但受控的方式，类可以通过这些方式扩展或重写继承的属性。
1. 类可以继承特质（trait）。将特质视为轻量级的类，允许以特定、有限的方式从多个超类继承。
1. （单例）对象（Object）是 Scala 中一种特殊的类。
  1. 它们不是上面所说的对象。记住，我们称那些为实例。

我们即将看到如何在 Scala 中创建一个类。

# 一个类的示例
创建一个 Scala 类的示例如下：

````scala
// WrapCounter counts up to a max value based on a bit size
class WrapCounter(counterBits: Int) {

  val max: Long = (1 << counterBits) - 1
  var counter = 0L
    
  def inc(): Long = {
    counter = counter + 1
    if (counter > max) {
        counter = 0
    }
    counter
  }
  println(s"counter created with max value $max")
}
````

这里包含了什么：
* ```class WrapCounter``` —— 这是 **WrapCounter** 的定义。
* ```(counterBits: Int)``` —— 创建一个 WrapCounter 需要一个整数参数，其名称恰当地暗示了它是计数器的位宽。
* 花括号（{}）界定一个代码块。大多数类使用代码块来定义变量、常量和方法（函数）。
* `val max: Long =` —— 该类包含一个成员变量 **max**，声明为 `Long` 类型，并在类创建时初始化。
* `(1 << counterBits) - 1` 计算 **counterBits** 位宽能容纳的最大值。由于 **max** 是用 `val` 创建的，所以它不能被改变。
* 创建了一个变量 **counter** 并初始化为 **0L**。**L** 表示 0 是一个 long 值；因此，**counter** 被推断为 Long 类型。
* **max** 和 **counter** 通常被称为类的*成员变量*。
* 定义了一个名为 **inc** 的类方法，它不接受参数并返回 **Long** 值。
* 方法 **inc** 的函数体是一个代码块，其中包含：
  * `counter = counter + 1` 使 **counter** 递增。
  * `if (counter > max) { counter = 0 }` 检查 counter 是否大于 **max** 值，如果是则将其重置为零。
  * `counter` —— 代码块的最后一行很重要。
    * 代码块最后一行表达的任何值都被视为该代码块的返回值。返回值可以被调用语句使用或忽略。
    * 这适用于非常广泛的情况；例如，由于 `if` `else` 语句用代码块定义其 true 和 false 子句，它可以返回值，即 `val result = if (10 * 10 > 90) "greater" else "lesser"` 会创建一个值为 "greater" 的 `val`。
  * 因此，在这种情况下，函数 **inc** 返回 **counter** 的值。
* `println(s"counter created with max value $max")` 向标准输出打印字符串。因为 **println** 直接在定义的代码块中，它是类初始化代码的一部分，每次创建该类的实例时都会运行，即打印出该字符串。
* 在这种情况下打印的字符串是一个*插值*字符串。
  * 第一个双引号前面的前导 **s** 将其标识为插值字符串。
  * 插值字符串在运行时处理。
  * **\$max** 被替换为 max 的值。
  * 如果 **\$** 后面跟着一个代码块，则该代码块中可以包含任意的 Scala 代码。
    * 例如，**`println(s"doubled max is ${max + max}")`**。
    * 该代码块的返回值将被插入到 `${...}` 的位置。
    * 如果返回值不是字符串，它将被转换为字符串；Scala 中几乎每个类或类型都有隐式转换为字符串的定义）。
  * 除非在调试，否则通常应该避免每次创建类的实例时都打印内容，以免淹没标准输出。

# 代码块
代码块由花括号界定。一个代码块可以包含零行或多行 Scala 代码。最后一行 Scala 代码成为代码块的返回值（可能被忽略）。没有行的代码块返回一个特殊的空值对象，称为 `Unit`。代码块在 Scala 中广泛使用：它们是类定义的主体，构成函数和方法定义，是 `if` 语句的子句，也是 `for` 和许多其他 Scala 运算符的主体。

## 参数化代码块
代码块可以接受参数。在类和方法定义的情况下，这些参数看起来像大多数传统编程语言中的参数。在下面的示例中，`c` 和 `s` 是代码块的参数。

````scala
// A one-line code block doesn't need to be enclosed in {}
def add1(c: Int): Int = c + 1

class RepeatString(s: String) {
  val repeatedString = s + s
}
````

**重要**：代码块还有另一种参数化方式。以下是一个示例。

````scala
val intList = List(1, 2, 3)
val stringList = intList.map { i =>
  i.toString
}
````

代码块被传递给类 List 的 `map` 方法。`map` 方法要求其代码块具有一个参数。代码块被调用于列表中的每个成员，代码块返回转换为 String 的该成员。Scala 对这种语法的变体几乎过于宽容。你可能会看到它以多种不同的方式编写。这种类型的代码块称为匿名函数，关于匿名函数的更多细节将在后续模块中提供。

这里的目标是帮助你在遇到不同的记法类型时识别它们。随着你使用 Scala，这些记法会变得更加舒适和熟悉。作者倾向于采用特定的风格，也存在个别语法场景下某种记法显得更自然。单行语句倾向于使用更简洁的形式。复杂的代码块通常具有更叙述性的外观。为了便于协作，建议浏览 [Scala 风格指南](http://docs.scala-lang.org/style/) 中的最佳实践。

# 命名参数和参数默认值
考虑以下方法定义。
```scala
def myMethod(count: Int, wrap: Boolean, wrapValue: Int = 24): Unit = { ... }
```
在调用该方法时，你通常会看到参数名称连同传入的值一起出现。
```scala
myMethod(count = 10, wrap = false, wrapValue = 23)
```
使用命名参数，你甚至可以以不同的顺序调用函数。
```scala
myMethod(wrapValue = 23, wrap = false, count = 10)
```
对于频繁调用的方法，参数顺序可能很明显。但对于不太常用的方法，特别是包含布尔参数时，在调用时包含参数名可以大大增加代码的可读性。如果方法有一长串相同类型的参数，使用名称也可以减少出错的机会。类定义的参数也使用这种命名参数方案（它们实际上只是类构造函数方法的参数）。

当某些参数有默认值（不需要被覆盖）时，调用者只需要传递（按名称）不使用默认值的特定参数。注意，参数 `wrapValue` 的默认值为 24。因此，

```scala
myMethod(wrap = false, count = 10)
```
将会像传入了 24 一样工作。
